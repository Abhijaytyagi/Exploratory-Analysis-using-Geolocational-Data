```python
food_data=food_data.dropna(axis=0)
food_data.head()
```

```python
important_features=['cook','eating_out','employment','ethnic_food','exercise','fruit_day','income','on_off_campus','pay_meal_out','sports','veggies_day']
food_data_final=food_data[important_features]
food_data_final.head()
```

```python
fig, ax = plt.subplots(figsize=(20,50 ))
sns.boxplot(data=food_data_final,ax=ax)
```

```python
x = food_data_final.iloc[:, [6]].values
```

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
```

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1,11 ), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=3)
y_kmeans2 = kmeans2.fit_predict(x)
print(y_kmeans2)

kmeans2.cluster_centers_
```
